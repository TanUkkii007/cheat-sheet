# 複素数

## 複素共役 (conjugate)

複素数zの複素共役を`z.real - z.imag`と定義し、$\bar{z}$と表記する。

## 絶対値

複素数zの絶対値$\|z\|$は三平方の定理より$|z|^2 = (z.\mathrm{real})^2 + (z.\mathrm{imag})^2$であり、$|z|^2 = z\cdot \bar{z}$である。

## オイラーの公式 (Euler’s formula)

任意の実数$\theta$に対し、$e^{i\theta}$は単位円上にあり偏角(argument)が$\theta$の点zに等しい。

$z = e^{i\theta}$

特に$\theta = \pi$のとき。

$e^{i\pi} + 1 = 0$

## 複素数の極表示

あらゆる複素数は極表示

$z = re^{\theta i}$

で表せる。ここで$\theta$はzの偏角、rは絶対値である。

## 回転

以下の関数はzを$\tau$だけ回転させる。

$f(z) = ze^{\tau i}$

導出

$re^{(\theta + \tau)i} = re^{\theta i}e^{\tau i}$

$ = ze^{\tau i}$

## 90度の回転

以下の関数はzを90度回転させる。

$f(z) = iz$

導出

$\frac{\pi}{2}$ラジアンの回転は、

$f(z) = ze^{i\pi/2}$

$f(z) = z(e^{i\pi})^{\frac{1}{2}}$

$e^{i\pi} = -1$より、

$f(z) = z\sqrt{-1}$

$f(z) = iz$

## 180度の回転

以下の関数はzを180度回転させる。

$f(z) = -z$

導出

$\pi$ラジアンの回転は、

$f(z) = ze^{i\pi}$

$e^{i\pi} = -1$より、

$f(z) = -z$
